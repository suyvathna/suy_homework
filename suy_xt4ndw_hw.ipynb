###2. task:

Create a House class that represents a house; attributes:


*   area
*   number of rooms
*   number of bathrooms
*   number of floors
*   has garage
*   has garden
*   any other attributes that you think is important



instance methods:

*   price calculation

There should be at least 1 default parameter

(4 points)

```python
class House():
  def __init__(self, age, area, n_bedrooms, n_bathrooms, n_kitchens, n_floors, has_garage = 0, has_garden = False, has_pool = False, has_fence = False):
    self.age = age
    self.area = area
    self.n_bedrooms = n_bedrooms
    self.n_bathrooms = n_bathrooms
    self.n_kitchens = n_kitchens
    self.n_floors = n_floors
    self.has_garage = has_garage
    self.has_garden = has_garden
    self.has_pool = has_pool
    self.has_fence = has_fence

  def __str__(self):
    return f"this {self.area}m2 house"

  def cal_price(self, rate_per_sqm = 200):
    return self.area * rate_per_sqm
```

```python
class Person():
  def __init__(self, name, age, n_children, n_cars, total_money, occupation, marital_stat = False):
    self.name = name
    self.age = age
    self.n_children = n_children
    self.n_cars = n_cars
    self.total_money = total_money
    self.occupation = occupation
    self.marital_stat = marital_stat

  def __str__(self):
    return f"{self.name}"

  # this method is used for checking if one can afford the house, so it should be a method in this Person class
  def affordability(self, house, rate_per_sqm=200):
    house_price = house.cal_price(rate_per_sqm)
    if (self.total_money *0.8) >= house_price:     # i think to buy a house, one should just spend at most 80% of their total money at a time
      print(f"{self.name} can afford the {house}")
    else:
      print(f"{house} is overbudget for {self.name}")

```

###A) (5 points)
 Create an Animal class and choose 3 animals that have some things in common (characteristics and behaviour)!

Write the Parent class and the derived classes of the species!

The Child classes should inherit some attributes and override some methods of the Parent class, but should also extend the state with own attributes in each subclass and extend the behaviour with own methods.

Decide yourself:

```python
# i choose a cat family which consists of 3 species: Lion, Tiger and Cheetah. They have common attibutes such as name, weight, habitat, and food.
# Each child class inherit the parent atributes except for Cheetah that i change the default attibute of weight to slightly less, while lion and tiger are considred the same.
# i put "hunt", "roar" and "string" method in Parent. However, i overwrite method "hunt" for each animal by their respetive strength and add a method "boom" to Cheetah for
# its speed.

# Parent class Animal
class Animal():
  def __init__(self, name, avg_weight = 200, habitat="wild", food = "meat"):# i found that to put argument without default value, they must have data type
    self.name = name
    self.avg_weight = avg_weight
    self.habitat = habitat
    self.food = food

  # hunt method
  def hunt(self):
    print(f"The {self.name} is very strong in hunting!")

  # roar method
  def roar(self):
    print(f"ROAR!!")

  # string method
  def __str__(self):
    return f"A {self.name} weights about {self.avg_weight} kg"


# child class Lion
class Lion(Animal):
  def __init__(self, name, avg_weight=200, habitat="wild", food = "meat", male_headfur = True):
    super().__init__(name, avg_weight, habitat, food)
    self.male_headfur = male_headfur

  # overwrite the existing hunt in Animal class
  def hunt(self):
    print(f"{self.name} is strong in using claw!")


# child class Tiger
class Tiger(Animal):
  def __init__(self, name, avg_weight=200, habitat="wild", food = "meat", stripe = True):
    super().__init__(name, avg_weight, habitat, food)
    self.stripe = stripe

  # overwrite parent's hunt method for Tiger
  def hunt(self):
    print(f"{self.name}'s teeth are powerful!")

# Another child class Cheetah
class Cheetah(Animal):
  def __init__(self, name, avg_weight=80, habitat="wild", food = "meat", top_speed = 117):
    super().__init__(name, avg_weight, habitat, food)
    self.top_speed = top_speed

  # overwrite parent's hunt method for Cheetah
  def hunt(self):
    print(f"A {self.name} hunts with their lightning fast speed!")

  # method for a Cheetah that runs so fast
  def boom(self):
    print(f"A {self.name} can reaches 60km/h in just 3 seconds and can reach up to {self.top_speed} km/h")
```

###B) (6 points)

Create classes and use inheritance to calculate the area of different shapes (specifically, circle and square, you can add triangle or anything else if you want).

Requirements:


1.	Create a base class called Shape that has a method named area().

This method should not have any implementation
2. Create at least two subclasses:

Circle — which takes a radius as a parameter

Square — which takes a side as a parameter

write the area-calculating function for both (or for all)

3. Create objects of both Circle and Square, and print their areas.

Extra: Store these objects in a list and use a loop to print the area of each shape.

```python
import math

class Shape():
  def __init__(self, name):
    self.name = name

  def area(self):
    pass

class Circle(Shape):
  def __init__(self, name, radius):
    super().__init__(name)
    self.radius = radius

  def area(self):
    return f"Area of {self.name} is {math.pi*self.radius**2}"

class Square(Shape):
  def __init__(self, name, side_length):
    super().__init__(name)
    self.side_length = side_length

  def area(self):
    return f"Area of {self.name} is {self.side_length**2}"

class Triangle(Shape):
  def __init__(self, name, base, height):
    super().__init__(name)
    self.base = base
    self.height = height

  def area(self):
    return f"Area of {self.name} is {(1/2)*self.base*self.height}"
```

```python
# test
c1 = Circle("c1",3)
c1.area()
```

```python
sqr1 = Square("sqr1",7)
print(sqr1.area())
```

```python
tri1 = Triangle("tri1",7,5)
tri1.area()
```

```python
# extra: store in lst
shapes = [c1,sqr1,tri1]

for s in shapes:
  print(s.area())
```
